```python
from typing import NamedTuple
from torch import Tensor
from torch_geometric.data import Data

class GraphGenout(NamedTuple):
    graph_batch: list[Data]
    avg_cosine_sim: Tensor
    std_cosine_sim: Tensor
```

####  POSSIBLE ISSUE:

Since the vit and deit were both trained with a the normal classification set up using cross entropy as loss, the resulting embeddings might not be embued with simmilar geometrical properties among samples (patches) sharing the same label. This can cause some patches to be either too similar between each other or too far away

```python
from torch_geometric.utils import dense_to_sparse
from torch_geometric.data import Data, Batch

def generate_sub_edge_index(adjacencies: Tensor):
    data_list = []
    for b in range(adjacencies.size(0)):
        edge_index, _ = dense_to_sparse(adjacencies[b])
        data = Data(
            edge_index=edge_index
        )
        data_list.append(data)

    # Combine all graphs into a single Batch
    # batch_graph = Batch.from_data_list(data_list)

    return data_list
```

```python
from torch import Tensor, bmm
from torch_geometric.utils import dense_to_sparse

def generate_connection(
    patches_emb: Tensor, 
    load_param: float,
    device: str = "cuda"
) -> GraphGenout:
    patches_emb = patches_emb / patches_emb.norm(dim=-1, keepdim=True)
    cosine_similarities: Tensor = bmm(
        input=patches_emb, 
        mat2=patches_emb.transpose(dim0=1, dim1=2)
    )

    cos_sim_sum: Tensor = cosine_similarities.sum(dim=[1, 2], keepdim=True)
    cos_sim_sum = cos_sim_sum.squeeze(dim=-1)
    avg_cosine_sim: Tensor = cos_sim_sum / (cosine_similarities.shape[1]**2)

    samples: Tensor = cosine_similarities.flatten(start_dim=1)
    std_cosine_sim: Tensor = (
        (samples - avg_cosine_sim)**2
    ).mean(dim=1, keepdim=True).sqrt()

    avg_cosine_sim = avg_cosine_sim.view(-1, 1, 1)
    std_cosine_sim = std_cosine_sim.view(-1, 1, 1)
    
    edge_mask: Tensor = (
        avg_cosine_sim - std_cosine_sim * load_param 
        <= cosine_similarities
    ) & (
        cosine_similarities
        <= avg_cosine_sim + std_cosine_sim * load_param
    )
    
    diagonal_mask = torch.tensor(
        [x for x in range(edge_mask.shape[1])],
        device=device
    )

    adjcency: Tensor = edge_mask.float()
    adjcency[:, diagonal_mask, diagonal_mask] = 0.

    graph_batch: list[Data] = generate_sub_edge_index(
        adjacencies=adjcency
    )

    return GraphGenout(
        graph_batch=graph_batch,
        avg_cosine_sim=avg_cosine_sim,
        std_cosine_sim=std_cosine_sim
    )
```

```python
def extract_topk_edges(explanation, graph_batch, top_k=10):
    edge_index = graph_batch.edge_index.cpu()
    edge_mask = explanation.edge_mask.detach().cpu()

    # Top-K edges by importance
    topk = min(top_k, edge_mask.shape[0])
    vals, idxs = torch.topk(edge_mask, topk)

    top_edges = []
    top_weights = []

    for eid, w in zip(idxs.tolist(), vals.tolist()):
        u, v = edge_index[:, eid].tolist()
        top_edges.append((u, v))
        top_weights.append(float(w))

    return top_edges, top_weights

def draw_edges(ax, G, pos):
    weights = nx.get_edge_attributes(G, "weight")
    max_w = max(weights.values()) if len(weights) > 0 else 1.0

    for (u, v), w in weights.items():
        # Normalize width for visibility
        width = 1.0 + 4.0 * (w / max_w)
        ax.plot(
            [pos[u][0], pos[v][0]],
            [pos[u][1], pos[v][1]],
            linewidth=width,
            color=(1 - w/max_w, 0, w/max_w),   # red/blue color scale
            alpha=0.8,
        )
```

```python
"""
gnn_explainer_full.py

Full pipeline:
 - load ViT backbone & GNN Lightning models (from checkpoints)
 - create patch embeddings from image
 - build PyG graph (uses your generate_connection)
 - run PyG Explainer (handles both 'object' and 'attributes' node masks)
 - produce heatmap overlay (224x224) and save
 - extract top-K nodes and edges, plot subgraph with actual image patches as node thumbnails

Requirements:
  - torch, torchvision
  - torch_geometric
  - opencv-python (cv2)
  - matplotlib
  - networkx
"""

import os
import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

from torch_geometric.explain import Explainer, GNNExplainer
from torch_geometric.data import Batch

# ==== replace these imports with your project's modules ====
from gnn import UltimateGraphApproach
from vit import VitClassifier
from utility import generate_connection, load_from_image_to_tensor, GraphGenout
# ============================================================


# ---------------------
# Helper functions
# ---------------------
def compute_node_importance_from_mask(node_mask: torch.Tensor) -> torch.Tensor:
    """
    Accepts node_mask from explanation (torch.Tensor).
    Handles both:
      - object masks: shape [num_nodes] -> return as is
      - attribute masks: shape [num_nodes, num_features] -> reduce to per-node score
    Returns: node_importance tensor shape [num_nodes]
    """
    if node_mask is None:
        raise ValueError("node_mask is None in explanation")

    node_mask = node_mask.detach().cpu()

    if node_mask.ndim == 1:
        # object-level mask already
        return node_mask.float()
    elif node_mask.ndim == 2:
        # attribute-level: aggregate across features into a single node score.
        # We use L2 norm (you can change to mean or sum).
        # shape [num_nodes, num_features] -> [num_nodes]
        return node_mask.norm(p=2, dim=1).float()
    else:
        raise ValueError(f"Unexpected node_mask ndim={node_mask.ndim}")


def build_patch_list_from_image(img: np.ndarray, patch_size: int = 16) -> list:
    """Slice image (H,W,3) into row-major list of patches (patch_size x patch_size)."""
    H, W = img.shape[:2]
    assert H % patch_size == 0 and W % patch_size == 0, "Image size must be divisible by patch size"
    grid_r = H // patch_size
    grid_c = W // patch_size
    patches = []
    for r in range(grid_r):
        for c in range(grid_c):
            p = img[r*patch_size:(r+1)*patch_size, c*patch_size:(c+1)*patch_size].copy()
            patches.append(p)
    return patches  # len = grid_r * grid_c


def save_heatmap_overlay(img: np.ndarray, node_importance: np.ndarray,
                         patch_size: int, out_path: str = "explanation_heatmap.png"):
    """
    node_importance: 1D array with length num_nodes (e.g. 196 or 197)
    If node_importance has extra CLS token (len = 197), we will ignore the last index for spatial mapping.
    """
    H, W = img.shape[:2]
    grid = H // patch_size
    spatial_nodes = grid * grid

    # If node_importance longer than spatial_nodes, assume final token(s) are non-spatial (CLS) and drop them.
    if len(node_importance) > spatial_nodes:
        node_importance = node_importance[:spatial_nodes]

    # create patch grid
    patch_grid = np.zeros((grid, grid), dtype=np.float32)
    arr = node_importance.copy()
    if arr.max() <= 0:
        # avoid division by zero later; small epsilon
        arr = arr + 1e-8

    # place into grid row-major
    patch_grid = arr.reshape((grid, grid))

    # normalize into [0,1]
    patch_grid = patch_grid - patch_grid.min()
    patch_grid = patch_grid / (patch_grid.max() + 1e-8)

    # upsample to image size
    heatmap_resized = cv2.resize(patch_grid, (W, H), interpolation=cv2.INTER_LINEAR)
    heatmap_uint8 = np.uint8(255 * heatmap_resized)
    heatmap_color = cv2.applyColorMap(heatmap_uint8, cv2.COLORMAP_JET)
    heatmap_color = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)

    # ensure img is uint8 RGB
    img_uint8 = img.copy()
    if img_uint8.dtype != np.uint8:
        img_uint8 = np.clip(img_uint8 * 255, 0, 255).astype(np.uint8)

    alpha = 0.45
    overlay = cv2.addWeighted(img_uint8, 1 - alpha, heatmap_color, alpha, 0)

    cv2.imwrite(out_path, cv2.cvtColor(overlay, cv2.COLOR_RGB2BGR))
    print(f"[INFO] saved heatmap overlay -> {out_path}")
    return out_path


def extract_topk_subgraph(explanation, graph_batch: Batch, top_k: int = 10):
    """
    Returns:
      - topk_nodes: list of node indices (spatial indices)
      - topk_scores: list of scores
      - important_edges: list of (u,v) among topk_nodes
      - important_edge_weights: list of floats
    """
    # node_mask might be object or attributes; handle both
    node_mask_raw = explanation.node_mask  # tensor
    node_importance = compute_node_importance_from_mask(node_mask_raw)  # [num_nodes]

    num_nodes = int(node_importance.shape[0])
    # if there is an extra token (e.g., CLS), patch_count = graph_batch.num_nodes could include it
    # We'll treat spatial_count = grid*grid, but keep node_importance for selecting topk.
    topk = min(top_k, num_nodes)
    topk_vals, topk_idx = torch.topk(node_importance, topk)
    topk_nodes = [int(x) for x in topk_idx.tolist()]
    topk_scores = [float(x) for x in topk_vals.tolist()]

    # edges
    edge_index = graph_batch.edge_index.detach().cpu()
    edge_mask = None
    if hasattr(explanation, "edge_mask") and explanation.edge_mask is not None:
        edge_mask = explanation.edge_mask.detach().cpu()
    else:
        # fallback: edge importance unknown -> treat equal weights
        edge_mask = torch.ones(edge_index.shape[1])

    important_edges = []
    important_edge_weights = []
    # iterate edge list (columns of edge_index)
    for eid, (u, v) in enumerate(edge_index.t().tolist()):
        if (u in topk_nodes) and (v in topk_nodes):
            important_edges.append((u, v))
            # handle length mismatch
            if eid < len(edge_mask):
                important_edge_weights.append(float(edge_mask[eid]))
            else:
                important_edge_weights.append(1.0)

    return topk_nodes, topk_scores, important_edges, important_edge_weights, node_importance.numpy()


def plot_subgraph_with_patches(
    img: np.ndarray,
    patches: list,
    topk_nodes: list,
    topk_scores: list,
    important_edges: list,
    important_edge_weights: list,
    node_importance_full: np.ndarray,
    patch_size: int = 16,
    save_path: str = "explanation_subgraph_patches.png"
):
    """
    Display subgraph where each node is represented by the image patch
    it corresponds to. Includes safety guards to prevent zero-sized patches.
    """

    # Build NetworkX graph
    G = nx.Graph()
    for n, score in zip(topk_nodes, topk_scores):
        G.add_node(n, importance=score)
    for (u, v), w in zip(important_edges, important_edge_weights):
        if u in G.nodes() and v in G.nodes():
            G.add_edge(u, v, weight=w)

    if len(G.nodes()) == 0:
        print("[WARN] No nodes to plot in subgraph.")
        return None

    # Layout
    pos = nx.spring_layout(G, seed=42)

    fig, ax = plt.subplots(figsize=(10, 10))

    # Draw edges
    if len(important_edge_weights) > 0:
        widths = [max(0.2, w * 3.0) for w in important_edge_weights]
        nx.draw_networkx_edges(G, pos, width=widths, alpha=0.8, edge_color="gray")

    # Draw image patches as nodes
    for n in G.nodes():

        # Safety check
        if not (0 <= n < len(patches)):
            print(f"[WARN] node {n}: index out of patch range")
            continue

        patch = patches[n]

        # Check: patch must be valid and non-empty
        if patch is None or patch.size == 0:
            print(f"[WARN] node {n}: empty patch — skipping")
            continue

        # Ensure dtype uint8
        if patch.dtype != np.uint8:
            patch_to_show = np.clip(patch * 255, 0, 255).astype(np.uint8)
        else:
            patch_to_show = patch

        # Check final patch resolution
        if patch_to_show.ndim != 3 or patch_to_show.shape[0] == 0 or patch_to_show.shape[1] == 0:
            print(f"[WARN] node {n}: unsafe patch size {patch_to_show.shape}")
            continue

        # Place the image as a node
        im = OffsetImage(patch_to_show, zoom=2.8)
        ab = AnnotationBbox(
            im,
            pos[n],
            frameon=True,
            pad=0.3,
            bboxprops=dict(edgecolor='black', facecolor='white', linewidth=1)
        )
        ax.add_artist(ab)

    # Optionally label nodes
    labels = {n: str(n) for n in G.nodes()}
    nx.draw_networkx_labels(G, pos, labels=labels, font_size=8)

    plt.title("Top-K Important Nodes (Patch Thumbnails)")
    plt.axis("off")
    plt.tight_layout()
    plt.savefig(save_path, dpi=200, bbox_inches="tight")
    plt.show()

    print(f"[INFO] saved subgraph plot -> {save_path}")
    return save_path

# ---------------------
# Main pipeline
# ---------------------
def run_full_explanation_pipeline(
    img_path: str,
    model_checkpoint_info: dict,
    vit_checkpoint_info: dict,
    patch_size: int = 16,
    image_size: int = 224,
    index_to_explain: int = 2,
    top_k: int = 8,
    explainer_epochs: int = 200,
    explainer_lr: float = 0.01,
):
    """
    model_checkpoint_info, vit_checkpoint_info: dicts with keys:
      - hparams_file
      - checkpoint_path

    Example:
      model_checkpoint_info = {
          "hparams_file": "FINAL_GNN_INNER_GAT\\GAT\\version_0\\hparams.yaml",
          "checkpoint_path": "FINAL_GNN_INNER_GAT\\GAT_wght\\weights.ckpt"
      }
    """

    device = "cuda" if torch.cuda.is_available() else "cpu"
    print("[INFO] device:", device)

    # -------------------------
    # 1) load image and create patches (and batch)
    # -------------------------
    batch_tensor, _ = load_from_image_to_tensor(img_path=img_path)
    # batch_tensor expected shape: [C,H,W] or what your Vit expects. We follow your earlier pipeline:
    batch_tensor = batch_tensor.unsqueeze(0).detach().to(device)

    # Load models (Lightning)
    model: UltimateGraphApproach = UltimateGraphApproach.load_from_checkpoint(
        hparams_file=model_checkpoint_info["hparams_file"],
        checkpoint_path=model_checkpoint_info["checkpoint_path"]
    )
    backbone: VitClassifier = VitClassifier.load_from_checkpoint(
        hparams_file=vit_checkpoint_info["hparams_file"],
        checkpoint_path=vit_checkpoint_info["checkpoint_path"]
    )
    model.eval()
    backbone.eval()
    model.to(device)
    backbone.to(device)

    # -------------------------
    # 2) compute patch embeddings then graph
    # -------------------------
    with torch.no_grad():
        patches_emb = backbone.predict_embedding(batch_tensor).detach().cpu()
    # Your generate_connection returns GraphGenout containing graph_batch list
    graph_out: GraphGenout = generate_connection(
        patches_emb=patches_emb,
        load_param=model.hparams.graph_load_param,
        device=device
    )

    # Create Batch (move to device for model)
    graph_batch = Batch.from_data_list(graph_out.graph_batch).to(device)
    # detach features
    if hasattr(graph_batch, "x") and graph_batch.x is not None:
        graph_batch.x = graph_batch.x.detach()

    # -------------------------
    # 3) build explainer
    # -------------------------
    explainer = Explainer(
        model=model.gnn,
        algorithm=GNNExplainer(epochs=explainer_epochs, lr=explainer_lr),
        explanation_type='model',
        # we prefer 'object' to get node-level masks. If attribute-level appears, we'll aggregate.
        node_mask_type='object',
        edge_mask_type='object',
        model_config=dict(
            mode='multiclass_classification',
            task_level='node',
            return_type='log_probs',
        ),
    )

    # -------------------------
    # 4) run explainer
    # -------------------------
    # Make sure inputs are detached and on CPU/GPU as expected
    x_input = graph_batch.x.detach().to(device)
    edge_index_input = graph_batch.edge_index.detach().to(device)

    explanation = explainer(x_input, edge_index_input, index=index_to_explain)
    print("[INFO] explanation produced")

    # -------------------------
    # 5) compute node_importance (numpy)
    # -------------------------
    node_importance_tensor = compute_node_importance_from_mask(explanation.node_mask)  # torch tensor
    node_importance = node_importance_tensor.cpu().numpy()  # numpy array

    # -------------------------
    # 6) load original image (RGB) and patches list
    # -------------------------
    # ----------------------------
# Force the image to be 224x224
# ----------------------------
    img_bgr = cv2.imread(img_path)
    if img_bgr is None:
        raise FileNotFoundError(f"Could not read image at: {img_path}")

    img_bgr = cv2.resize(img_bgr, (image_size, image_size), interpolation=cv2.INTER_AREA)
    img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)


    patches = build_patch_list_from_image(img, patch_size=patch_size)  # list length grid*grid

    # -------------------------
    # 7) heatmap overlay
    # -------------------------
    heatmap_path = "explanation_heatmap.png"
    save_heatmap_overlay(img, node_importance, patch_size, out_path=heatmap_path)

    # -------------------------
    # 8) extract subgraph top-k and plot patches
    # -------------------------
    topk_nodes, topk_scores, important_edges, important_edge_weights, _ = extract_topk_subgraph(
        explanation, graph_batch, top_k=top_k
    )

    # Safety: if some topk nodes exceed patches count (e.g., CLS token), drop them
    grid_nodes = (image_size // patch_size) * (image_size // patch_size)
    filtered_topk_nodes = []
    filtered_topk_scores = []
    for n, s in zip(topk_nodes, topk_scores):
        if 0 <= n < grid_nodes:
            filtered_topk_nodes.append(n)
            filtered_topk_scores.append(s)
        else:
            print(f"[INFO] skipping non-spatial node {n} (likely CLS or similar)")

    # Filter edges to those connecting filtered_topk_nodes
    filtered_edges = []
    filtered_edge_weights = []
    for (u, v), w in zip(important_edges, important_edge_weights):
        if u in filtered_topk_nodes and v in filtered_topk_nodes:
            filtered_edges.append((u, v))
            filtered_edge_weights.append(w)

    plot_subgraph_with_patches(
        img=img,
        patches=patches,
        topk_nodes=filtered_topk_nodes,
        topk_scores=filtered_topk_scores,
        important_edges=filtered_edges,
        important_edge_weights=filtered_edge_weights,
        node_importance_full=node_importance,
        patch_size=patch_size,
        save_path="explanation_subgraph_patches.png"
    )

    return {
        "heatmap_path": heatmap_path,
        "subgraph_path": "explanation_subgraph_patches.png",
        "explanation": explanation,
        "node_importance": node_importance
    }


# ---------------------
# Example invocation
# ---------------------
if __name__ == "__main__":
    img_path = "dataset\\test\\Egyptian\\Frag_107.png"

    model_ckpt = {
        "hparams_file": "FINAL_GNN_INNER_GAT\\GAT\\version_0\\hparams.yaml",
        "checkpoint_path": "FINAL_GNN_INNER_GAT\\GAT_wght\\weights.ckpt"
    }
    vit_ckpt = {
        "hparams_file": "final_VIT\\FULL_VIT_TEST_logs\\FINAL_VIT_csv_extrapolated\\version_0\\hparams_extrapolated.yaml",
        "checkpoint_path": "final_VIT\\FINAL_VIT_CHKT\\weights_extrapolated.ckpt"
    }

    results = run_full_explanation_pipeline(
        img_path=img_path,
        model_checkpoint_info=model_ckpt,
        vit_checkpoint_info=vit_ckpt,
        patch_size=16,
        image_size=224,
        index_to_explain=2,
        top_k=8,
        explainer_epochs=200,
        explainer_lr=0.01,
    )

    print("[DONE] heatmap:", results["heatmap_path"])
    print("[DONE] subgraph:", results["subgraph_path"])
```

```python
import numpy as np
import cv2

def save_heatmap_overlay_true(
    img: np.ndarray,
    node_importance: np.ndarray,
    patch_size: int,
    out_path: str = "explanation_heatmap.png",
    alpha: float = 0.45,
):
    """
    img: H x W x 3 RGB image
    node_importance: (N,) where N = 1 + num_patches (CLS + patches)
    patch_size: e.g. 16
    """

    H, W = img.shape[:2]

    # number of spatial patches
    spatial_nodes = (H // patch_size) * (W // patch_size)

    # drop CLS token if present
    if len(node_importance) == spatial_nodes + 1:
        node_importance = node_importance[1:]
    elif len(node_importance) != spatial_nodes:
        raise ValueError(
            f"node_importance has length {len(node_importance)}, "
            f"expected {spatial_nodes} or {spatial_nodes + 1}"
        )

    # infer grid size (e.g. 16x16)
    grid_h = H // patch_size
    grid_w = W // patch_size

    # reshape to patch grid
    patch_grid = node_importance.reshape(grid_h, grid_w).astype(np.float32)

    # normalize to [0, 1]
    patch_grid -= patch_grid.min()
    patch_grid /= (patch_grid.max() + 1e-8)

    # upsample to image resolution
    heatmap = cv2.resize(
        patch_grid, (W, H), interpolation=cv2.INTER_NEAREST
    )

    # smooth edges
    heatmap = cv2.GaussianBlur(
        heatmap, ksize=(0, 0), sigmaX=patch_size / 2
    )

    heatmap_uint8 = np.uint8(255 * heatmap)
    heatmap_color = cv2.applyColorMap(
        heatmap_uint8, cv2.COLORMAP_TURBO
    )
    heatmap_color = cv2.cvtColor(
        heatmap_color, cv2.COLOR_BGR2RGB
    )

    # ensure image is uint8
    if img.dtype != np.uint8:
        img = np.clip(img * 255, 0, 255).astype(np.uint8)

    overlay = cv2.addWeighted(
        img, 1 - alpha, heatmap_color, alpha, 0
    )

    cv2.imwrite(out_path, cv2.cvtColor(overlay, cv2.COLOR_RGB2BGR))
    print(f"[INFO] saved heatmap overlay -> {out_path}")

    return out_path
```

```python
"""
gnn_explainer_full.py

Full pipeline:
 - load ViT backbone & GNN Lightning models (from checkpoints)
 - create patch embeddings from image
 - build PyG graph (uses your generate_connection)
 - run PyG Explainer (handles both 'object' and 'attributes' node masks)
 - produce heatmap overlay (224x224) and save
 - extract top-K nodes and edges, plot subgraph with actual image patches as node thumbnails

Requirements:
  - torch, torchvision
  - torch_geometric
  - opencv-python (cv2)
  - matplotlib
  - networkx
"""

import os
import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

from torch_geometric.explain import Explainer, GNNExplainer
from torch_geometric.data import Batch

# ==== replace these imports with your project's modules ====
from gnn import UltimateGraphApproach
from vit import VitClassifier
from utility import generate_connection, load_from_image_to_tensor, GraphGenout
# ============================================================


# ---------------------
# Helper functions
# ---------------------
def compute_node_importance_from_mask(node_mask: torch.Tensor) -> torch.Tensor:
    """
    Accepts node_mask from explanation (torch.Tensor).
    Handles both:
      - object masks: shape [num_nodes] -> return as is
      - attribute masks: shape [num_nodes, num_features] -> reduce to per-node score
    Returns: node_importance tensor shape [num_nodes]
    """
    if node_mask is None:
        raise ValueError("node_mask is None in explanation")

    node_mask = node_mask.detach().cpu()

    if node_mask.ndim == 1:
        # object-level mask already
        return node_mask.float()
    elif node_mask.ndim == 2:
        # attribute-level: aggregate across features into a single node score.
        # We use L2 norm (you can change to mean or sum).
        # shape [num_nodes, num_features] -> [num_nodes]
        return node_mask.norm(p=2, dim=1).float()
    else:
        raise ValueError(f"Unexpected node_mask ndim={node_mask.ndim}")


def build_patch_list_from_image(img: np.ndarray, patch_size: int = 16) -> list:
    """Slice image (H,W,3) into row-major list of patches (patch_size x patch_size)."""
    H, W = img.shape[:2]
    assert H % patch_size == 0 and W % patch_size == 0, "Image size must be divisible by patch size"
    grid_r = H // patch_size
    grid_c = W // patch_size
    patches = []
    for r in range(grid_r):
        for c in range(grid_c):
            p = img[r*patch_size:(r+1)*patch_size, c*patch_size:(c+1)*patch_size].copy()
            patches.append(p)
    return patches  # len = grid_r * grid_c


def save_heatmap_overlay(img: np.ndarray, node_importance: np.ndarray,
                         patch_size: int, out_path: str = "explanation_heatmap.png"):
    """
    node_importance: 1D array with length num_nodes (e.g. 196 or 197)
    If node_importance has extra CLS token (len = 197), we will ignore the last index for spatial mapping.
    """
    H, W = img.shape[:2]
    grid = H // patch_size
    spatial_nodes = grid * grid

    # If node_importance longer than spatial_nodes, assume final token(s) are non-spatial (CLS) and drop them.
    if len(node_importance) > spatial_nodes:
        node_importance = node_importance[:spatial_nodes]

    # create patch grid
    patch_grid = np.zeros((grid, grid), dtype=np.float32)
    arr = node_importance.copy()
    if arr.max() <= 0:
        # avoid division by zero later; small epsilon
        arr = arr + 1e-8

    # place into grid row-major
    patch_grid = arr.reshape((grid, grid))

    # normalize into [0,1]
    patch_grid = patch_grid - patch_grid.min()
    patch_grid = patch_grid / (patch_grid.max() + 1e-8)

    # upsample to image size
    heatmap_resized = cv2.resize(patch_grid, (W, H), interpolation=cv2.INTER_LINEAR)
    heatmap_uint8 = np.uint8(255 * heatmap_resized)
    heatmap_color = cv2.applyColorMap(heatmap_uint8, cv2.COLORMAP_JET)
    heatmap_color = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)

    # ensure img is uint8 RGB
    img_uint8 = img.copy()
    if img_uint8.dtype != np.uint8:
        img_uint8 = np.clip(img_uint8 * 255, 0, 255).astype(np.uint8)

    alpha = 0.45
    overlay = cv2.addWeighted(img_uint8, 1 - alpha, heatmap_color, alpha, 0)

    cv2.imwrite(out_path, cv2.cvtColor(overlay, cv2.COLOR_RGB2BGR))
    print(f"[INFO] saved heatmap overlay -> {out_path}")
    return out_path


def extract_topk_subgraph(explanation, graph_batch: Batch, top_k: int = 10):
    """
    Returns:
      - topk_nodes: list of node indices (spatial indices)
      - topk_scores: list of scores
      - important_edges: list of (u,v) among topk_nodes
      - important_edge_weights: list of floats
    """
    # node_mask might be object or attributes; handle both
    node_mask_raw = explanation.node_mask  # tensor
    node_importance = compute_node_importance_from_mask(node_mask_raw)  # [num_nodes]

    num_nodes = int(node_importance.shape[0])
    # if there is an extra token (e.g., CLS), patch_count = graph_batch.num_nodes could include it
    # We'll treat spatial_count = grid*grid, but keep node_importance for selecting topk.
    topk = min(top_k, num_nodes)
    topk_vals, topk_idx = torch.topk(node_importance, topk)
    topk_nodes = [int(x) for x in topk_idx.tolist()]
    topk_scores = [float(x) for x in topk_vals.tolist()]

    # edges
    edge_index = graph_batch.edge_index.detach().cpu()
    edge_mask = None
    if hasattr(explanation, "edge_mask") and explanation.edge_mask is not None:
        edge_mask = explanation.edge_mask.detach().cpu()
    else:
        # fallback: edge importance unknown -> treat equal weights
        edge_mask = torch.ones(edge_index.shape[1])

    important_edges = []
    important_edge_weights = []
    # iterate edge list (columns of edge_index)
    for eid, (u, v) in enumerate(edge_index.t().tolist()):
        if (u in topk_nodes) and (v in topk_nodes):
            important_edges.append((u, v))
            # handle length mismatch
            if eid < len(edge_mask):
                important_edge_weights.append(float(edge_mask[eid]))
            else:
                important_edge_weights.append(1.0)

    return topk_nodes, topk_scores, important_edges, important_edge_weights, node_importance.numpy()


def plot_subgraph_with_patches(
    img: np.ndarray,
    patches: list,
    topk_nodes: list,
    topk_scores: list,
    important_edges: list,
    important_edge_weights: list,
    node_importance_full: np.ndarray,
    patch_size: int = 16,
    save_path: str = "explanation_subgraph_patches.png"
):
    """
    Display subgraph where each node is represented by the image patch
    it corresponds to. Includes safety guards to prevent zero-sized patches.
    """

    # Build NetworkX graph
    G = nx.Graph()
    for n, score in zip(topk_nodes, topk_scores):
        G.add_node(n, importance=score)
    for (u, v), w in zip(important_edges, important_edge_weights):
        if u in G.nodes() and v in G.nodes():
            G.add_edge(u, v, weight=w)

    if len(G.nodes()) == 0:
        print("[WARN] No nodes to plot in subgraph.")
        return None

    # Layout
    pos = nx.spring_layout(G, seed=42)

    fig, ax = plt.subplots(figsize=(10, 10))

    # # Draw edges
    # if len(important_edge_weights) > 0:
    #     widths = [max(0.2, w * 3.0) for w in important_edge_weights]
    #     nx.draw_networkx_edges(G, pos, width=widths, alpha=0.8, edge_color="gray")

    draw_edges(ax, G, pos)
    # Draw image patches as nodes
    for n in G.nodes():

        # Safety check
        if not (0 <= n < len(patches)):
            print(f"[WARN] node {n}: index out of patch range")
            continue

        patch = patches[n]

        # Check: patch must be valid and non-empty
        if patch is None or patch.size == 0:
            print(f"[WARN] node {n}: empty patch — skipping")
            continue

        # Ensure dtype uint8
        if patch.dtype != np.uint8:
            patch_to_show = np.clip(patch * 255, 0, 255).astype(np.uint8)
        else:
            patch_to_show = patch

        # Check final patch resolution
        if patch_to_show.ndim != 3 or patch_to_show.shape[0] == 0 or patch_to_show.shape[1] == 0:
            print(f"[WARN] node {n}: unsafe patch size {patch_to_show.shape}")
            continue

        # Place the image as a node
        im = OffsetImage(patch_to_show, zoom=2.8)
        ab = AnnotationBbox(
            im,
            pos[n],
            frameon=True,
            pad=0.3,
            bboxprops=dict(edgecolor='black', facecolor='white', linewidth=1)
        )
        ax.add_artist(ab)

    # Optionally label nodes
    labels = {n: str(n) for n in G.nodes()}
    nx.draw_networkx_labels(G, pos, labels=labels, font_size=8)

    plt.title("Top-K Important Nodes (Patch Thumbnails)")
    plt.axis("off")
    plt.tight_layout()
    plt.savefig(save_path, dpi=200, bbox_inches="tight")
    plt.show()

    print(f"[INFO] saved subgraph plot -> {save_path}")
    return save_path

# ---------------------
# Main pipeline
# ---------------------
def run_full_explanation_pipeline(
    img_path: str,
    model_checkpoint_info: dict,
    vit_checkpoint_info: dict,
    heatmap_path: str, 
    patch_size: int = 16,
    image_size: int = 224,
    index_to_explain: int = 2,
    top_k: int = 20,
    explainer_epochs: int = 200,
    explainer_lr: float = 0.01,
):
    """
    model_checkpoint_info, vit_checkpoint_info: dicts with keys:
      - hparams_file
      - checkpoint_path

    Example:
      model_checkpoint_info = {
          "hparams_file": "FINAL_GNN_INNER_GAT\\GAT\\version_0\\hparams.yaml",
          "checkpoint_path": "FINAL_GNN_INNER_GAT\\GAT_wght\\weights.ckpt"
      }
    """

    device = "cuda" if torch.cuda.is_available() else "cpu"
    print("[INFO] device:", device)

    # -------------------------
    # 1) load image and create patches (and batch)
    # -------------------------
    batch_tensor, _ = load_from_image_to_tensor(img_path=img_path)
    # batch_tensor expected shape: [C,H,W] or what your Vit expects. We follow your earlier pipeline:
    batch_tensor = batch_tensor.unsqueeze(0).detach().to(device)

    # Load models (Lightning)
    model: UltimateGraphApproach = UltimateGraphApproach.load_from_checkpoint(
        hparams_file=model_checkpoint_info["hparams_file"],
        checkpoint_path=model_checkpoint_info["checkpoint_path"]
    )
    backbone: VitClassifier = VitClassifier.load_from_checkpoint(
        hparams_file=vit_checkpoint_info["hparams_file"],
        checkpoint_path=vit_checkpoint_info["checkpoint_path"]
    )
    model.eval()
    backbone.eval()
    model.to(device)
    backbone.to(device)

    # -------------------------
    # 2) compute patch embeddings then graph
    # -------------------------
    with torch.no_grad():
        patches_emb = backbone.predict_embedding(batch_tensor).detach().cpu()
    # Your generate_connection returns GraphGenout containing graph_batch list
    graph_out: GraphGenout = generate_connection(
        patches_emb=patches_emb,
        load_param=model.hparams.graph_load_param,
        device=device
    )

    # Create Batch (move to device for model)
    graph_batch = Batch.from_data_list(graph_out.graph_batch).to(device)
    # detach features
    if hasattr(graph_batch, "x") and graph_batch.x is not None:
        graph_batch.x = graph_batch.x.detach()

    # -------------------------
    # 3) build explainer
    # -------------------------
    explainer = Explainer(
        model=model.gnn,
        algorithm=GNNExplainer(epochs=explainer_epochs, lr=explainer_lr),
        explanation_type='model',
        # we prefer 'object' to get node-level masks. If attribute-level appears, we'll aggregate.
        node_mask_type='object',
        edge_mask_type='object',
        model_config=dict(
            mode='multiclass_classification',
            task_level='node',
            return_type='log_probs',
        ),
    )

    # -------------------------
    # 4) run explainer
    # -------------------------
    # Make sure inputs are detached and on CPU/GPU as expected
    x_input = graph_batch.x.detach().to(device)
    edge_index_input = graph_batch.edge_index.detach().to(device)

    explanation = explainer(x_input, edge_index_input, index=index_to_explain)
    print("[INFO] explanation produced")

    # -------------------------
    # 5) compute node_importance (numpy)
    # -------------------------
    node_importance_tensor = compute_node_importance_from_mask(explanation.node_mask)  # torch tensor
    node_importance = node_importance_tensor.cpu().numpy()  # numpy array

    # -------------------------
    # 6) load original image (RGB) and patches list
    # -------------------------
    # ----------------------------
# Force the image to be 224x224
# ----------------------------
    img_bgr = cv2.imread(img_path)
    if img_bgr is None:
        raise FileNotFoundError(f"Could not read image at: {img_path}")

    img_bgr = cv2.resize(img_bgr, (image_size, image_size), interpolation=cv2.INTER_AREA)
    img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)


    patches = build_patch_list_from_image(img, patch_size=patch_size)  # list length grid*grid

    # -------------------------
    # 7) heatmap overlay
    # -------------------------
    save_heatmap_overlay_true(img, node_importance, patch_size, out_path=heatmap_path)

    # -------------------------
    # 8) extract subgraph top-k and plot patches
    # -------------------------
    topk_nodes, topk_scores, important_edges, important_edge_weights, _ = extract_topk_subgraph(
        explanation, graph_batch, top_k=top_k
    )

    # Safety: if some topk nodes exceed patches count (e.g., CLS token), drop them
    grid_nodes = (image_size // patch_size) * (image_size // patch_size)
    filtered_topk_nodes = []
    filtered_topk_scores = []
    for n, s in zip(topk_nodes, topk_scores):
        if 0 <= n < grid_nodes:
            filtered_topk_nodes.append(n)
            filtered_topk_scores.append(s)
        else:
            print(f"[INFO] skipping non-spatial node {n} (likely CLS or similar)")

    # Filter edges to those connecting filtered_topk_nodes
    filtered_edges = []
    filtered_edge_weights = []
    for (u, v), w in zip(important_edges, important_edge_weights):
        if u in filtered_topk_nodes and v in filtered_topk_nodes:
            filtered_edges.append((u, v))
            filtered_edge_weights.append(w)

    # plot_subgraph_with_patches(
    #     img=img,
    #     patches=patches,
    #     topk_nodes=filtered_topk_nodes,
    #     topk_scores=filtered_topk_scores,
    #     important_edges=filtered_edges,
    #     important_edge_weights=filtered_edge_weights,
    #     node_importance_full=node_importance,
    #     patch_size=patch_size,
    #     save_path="explanation_subgraph_patches3.png"
    # )

    return {
        "heatmap_path": heatmap_path,
        "subgraph_path": "explanation_subgraph_patches2.png",
        "explanation": explanation,
        "node_importance": node_importance
    }


# ---------------------
# Example invocation
# ---------------------
```
